```python
x = df.drop("species",axis=1)
y = df['species']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model_1 = KNeighborsClassifier(metric='minkowski')
model_1.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score,f1_score,classification_report
```

```python
model_1.score(X_test,y_test)
```

```python
y_predict = model_1.predict(X_test)
accuracy_score(y_predict,y_test)
```

```python
from sklearn.svm import SVC
```

```python
model_SC = SVC(gamma='auto')
model_SC.fit(X_train,y_train)
```

```python
model_SC.score(X_test,y_test)
```

```python
y_predict_Sv = model_1.predict(X_test)
accuracy_score(y_predict_Sv,y_test)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
classfier = GridSearchCV((model_SC),{
    "C":[1,3,5,8,10,30,40],
    "kernel":['linear','rbf']
},cv=7,return_train_score=False)
```

```python
classfier.fit(x,y)
```

```python
classfier.cv_results_
```

```python
result = pd.DataFrame(classfier.cv_results_)
```
